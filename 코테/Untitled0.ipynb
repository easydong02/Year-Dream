## 프린터

```python
#최종
def solution(priorities, location):
    # 딕셔너리로 만듦
    dict_ = {i : priorities[i] for i in range(len(priorities))}

    # 원하는 문서 인쇄 순서
    answer=0

    # 최종 인쇄 순서 딕셔너리
    result ={}

    while True:
        # 종료 조건
        if len(dict_)==0:
            break

        #  처음, 마지막 키값
        key_f,key_l = next(iter(dict_)), len(priorities)

        #유배 작업 체크 변수
        check = False

        #첫번째거를 나머지비교
        first = dict_[key_f]
        for a,b in dict_.items():
            if b> first:
                check=True
        
        # 유배 작업
        if check:
            temp_v = dict_.pop(key_f)
            dict_[key_f +len(priorities)] = temp_v
            
        # 더 큰 거 없으면 인쇄
        else:
            temp=dict_.pop(key_f)
            result[key_f%len(priorities)] = temp

    #원하는 문서 출력 순서
    for a,b in result.items():
        answer +=1
        if a==location:
            return answer
    
solution([1, 1, 9, 1, 1, 1],0)
```

## 다리를 지나가는 트럭

* 1. 차가 다리에 있는 시간 = bridge_length
* 2. 완전히 건너는 시간 = bridge_length + 1
* 3. 다리위에 차 최대 조건 = (truck_weigth * truck_cnt <= weight) & (truck_cnt <= bridge_length)

```python
# 초안
def solution(bridge_length, weight, truck_weights):
    on_bridge = []                                  # 다리 위에 있는 차 리스트
    on_time = [0 for i in range(len(truck_weights))]# 각각의 차 시간
    off_bridge= []                                  # 나간 차 리스트
    tic=0                                           # 전체 시간
    check = len(truck_weights)
    while True:
        # 차 한대만 있을때
        if check == 1:
            tic = bridge_length + 1
            break


        # 차가 들어가고 전체 이동
        if sum(on_bridge) <=weight:
            if len(truck_weights) >0 :
                on_bridge.append(truck_weights.pop(0))
                for i in range(len(on_bridge)):
                    on_time[i] +=1
            tic +=1
    
        
        # 대기열 없을 때 빨리 끝내기
        if len(truck_weights) ==0:
            if len(on_bridge) ==1:
                tic +=(bridge_length-1)
                break
            tic += bridge_length
            break

        # 대기열 있는데 차 한 대 나갈 때
        else:
            if on_time[0] == bridge_length-1:
                on_time.pop(0)
                off_bridge.append(on_bridge.pop(0))
                tic +=1
        

    
    answer = tic
    return answer

solution(2, 10, [7, 4, 5, 6])
```

```python
import numpy as np

# 첫 조건 자리
def solution(bridge_length, weight, truck_weights):
    on_bridge = []
    on_time = np.array([])
    tic =0


    while True:
        # 다리에 자리 있으면 차 반입 시도
        if  len(on_bridge) < bridge_length :
            
            # 무게되면 차 반입 + 나갈 수 있으면 차 반출
            if sum(on_bridge) + truck_weights[0] <=weight :
                on_bridge.append(truck_weights.pop(0))
                on_time = np.append(on_time, 0)
                if on_time[0] >=bridge_length :
                    on_bridge.pop(0)
                    on_time = np.delete(on_time,0)

            # 그렇지 않고 선입차 나갈 수 있으면 차 반출 + 대기열 차 반입
            elif on_time[0] >=bridge_length :
                on_bridge.pop(0)
                on_time = np.delete(on_time,0)

                if sum(on_bridge) + truck_weights[0] <=weight :
                    on_bridge.append(truck_weights.pop(0))
                    on_time = np.append(on_time, 0)

        # 그렇지 않으면 다리 꽉참 -> 맨 앞 하나 뺌 + (무게 되면)대기열에서 하나 들임
        else:
            on_bridge.pop(0)
            on_time = np.delete(on_time,0)
            if sum(on_bridge) + truck_weights[0] <=weight :
                on_bridge.append(truck_weights.pop(0))
                on_time = np.append(on_time, 0)

        # tic은 다리에 차를 넣어도, 넣지 않아도 차들은 움직이고 시간은 흐름
        tic +=1
        on_time +=1
        
        # 합산
        if len(truck_weights) ==0:
            tic += bridge_length
            break
        

    answer=tic
    return answer


solution( 2, 10, [7, 4, 5, 6] 	)
```

## 주식가격

* 1. 각 가격보다 낮은 가격의 인덱스를 찾아서 인덱스 뺀다.

```python
# numpy 사용했을 때
import numpy as np

def solution(prices):
    prices_ = np.array(prices)
    answer = []
    for i in range(len(prices_)):
        idx = np.where(prices_[i:]<prices_[i])  # 이게 효율성 탈락원인
        
        if len(idx[0]) ==0: # 최소값이면
            answer.append(len(prices_)-1 -i)
        else:
            answer.append(int(idx[0][0])) # 자기보다 작아지는 처음 값의 인덱스

    return answer

solution([1, 2, 3, 2, 3])
```

## 위장

```python
# 모든 수를 직접 구했을 때
from collections import Counter
from itertools import combinations
def multi(list_):
    result =0
    for i in list_:
        temp=1
        for j in i:
            temp *=j
        result +=temp
    return result


def solution(clothes):
    answer = 0
    parts= [i[1] for i in clothes]
    parts_ = list(Counter(parts).values())
    if len(parts_)>1:
        for i in range(1,len(parts_)+1):
            answer += multi(list(combinations(parts_,i)))
    else:
        answer = parts_[0]

    return answer


solution([["crowmask", "face"], ["bluesunglasses", "face"], ["smoky_makeup", "face"]])
```

```python
# 공식
from collections import Counter


def solution(clothes):
    parts= [i[1] for i in clothes]
    parts_ = list(Counter(parts).values())
    temp=1
    for i in parts_:
        temp *= (i+1)

    return temp-1


solution([["crowmask", "face"], ["bluesunglasses", "face"], ["smoky_makeup", "face"]])
```

## 더 맵게

```python
#  PriorityQueue 사용 ->효율성 실패
from queue import PriorityQueue


def solution(scoville, K):
    answer = 0
    result = PriorityQueue()
    for i in scoville:
        result.put(i)
    while True:
        temp = result.get()
        if temp < K:
            result.put(temp+ (result.get()*2))
            answer +=1
        else:
            break
    
    return answer


solution([1, 2, 3, 9, 10, 12],7)
```

```python
# Heapq 사용
import heapq


def solution(scoville, K):
    answer = 0
    result = []
    for i in scoville:
        heapq.heappush(result,i)

    while True:
        temp=heapq.heappop(result)
        

        if temp <K:
            if len(result) ==0:
                return -1
            heapq.heappush(result, temp+ (heapq.heappop(result)*2))
            answer +=1


        else:
            break
    
    return answer


solution([10, 12],70)
```

## 소수 찾기


```python
import itertools

# 소수 인지 확인하는 함수 
def find(num):
    if num ==1:
        return False
    for i in range(2,num):
        if (num %i == 0):
            return False
    return True

def solution(numbers):
    numbers_ = [i for i in numbers]
    result = []
    answer = 0
    for i in range(1,len(numbers_)+1):
        nPr=list(itertools.permutations(numbers_,i))
        for j in nPr:
            num = int("".join(j))
            if num != 0:
                if i==1:
                    if (num not in result) :
                        result.append(num)
                else:
                    if num not in result :
                        result.append(num)
    
    for i in result:
        if find(i):
            answer +=1

    return answer


solution('011')
```

## 카펫

- x * y = brown + yellow

- x + y = brown/2 +2

- x,y >=3





```python
def solution(brown, yellow):
    answer = []
    xy= brown + yellow
    x_plus_y = int((brown/2) +2)

    temp =3

    while True:
        x = x_plus_y -temp
        y = temp
        if (xy%y ==0) and ( xy==x*y):
            answer.append(x)
            answer.append(y)
            break
        temp +=1
    return answer

solution(24,24)
```

## 타겟넘버

- 다 더한 수에서 타겟넘버가 되려면 필요한 마이너스 숫자를 구한다?

```python
from itertools import combinations

def solution(numbers, target):
    answer = 0
    numbers_= [2*i for i in numbers]
    make = sum(numbers) - target
    for i in range(1,len(numbers_)+1):
        result =  [sum(e) for e in  list(combinations(numbers_,i))]
        for j in result:
            if j ==make:
                answer +=1


    return answer

solution([4, 1, 2, 1],4)
```

## 큰 수 만들기

- 순서 변경 X
- 첫 숫자 정하기
- 남은 숫자에서 가장 큰 수 만들기 (처음부터 돌면서 뒤에 숫자보다 작으면 그 숫자 삭제)

```python
#시간 초과
def solution(number, k):
    number =list(number)
    while k>0:
        cnt = 0
        for i in range(len(number)-1):
            cnt +=1
            if number[i]<number[i+1]:
                del number[i]
                break
            if cnt == len(number)-1:
                del number[-1]
        k -=1

    return "".join(number)


solution("9998881234",4)
```

## 구명보트

- 첫번째 사람부터 리미트에서 몸무게 빼가면서 검색

```python
from collections import deque

def solution(people, limit):
    answer = 0
    people_ = deque(sorted(people))
    while people_ :
        if len(people_)==1:
            answer +=1
            break
        
        if people_[0] +people_[-1] >limit:
            people_.pop()
            answer +=1
        else:
            people_.pop()
            people_.popleft()
            answer +=1

    return answer

solution([70, 50, 80, 50],100)
```
