```python
import numpy as np
```

$$
y = e^{2x+1}
$$

```python
def add(x, a):
    # WRITE CODE HERE [+]
    return x+a , 1

def mul(x, a):
    # WRITE CODE HERE [+]
    return a*x, a
    

def exp(x):
    # WRITE CODE HERE [+]
    return np.exp(x),np.exp(x)
```

```python
x0 = 1.5

# WRITE CODE HERE [+]
a, dx0 = mul(x0,2)

b, da = add(a,1)

y, db = exp(b)

print(f"f({x0})={y:.3f}")
print(f"f'({x0})={dx0*da*db:.3f}")
```

$$
y = \left( \frac{\sqrt{x}}{2} - 1 \right)^{-10}
$$

```python
import sympy as S
```

```python
x = S.Symbol('x')

y = (S.sqrt(x)/2 -1)**(-10)
y
```

```python
dy = S.diff(y, x)
dy
```

```python
x0 = 1
print(f"f({x0})={y.subs({x:x0})}" )
print(f"f'({x0})={dy.subs({x:x0})}" )
```

```python
def pow(x, n):
    # WRITE CODE HERE [+]
    return x**n, n*x**(n-1)
```

```python
x0 = 1

# WRITE CODE HERE [+]
a, dx0 = pow(x0,(1/2))
b, da  = mul(a,(1/2))
c, db = add(b,-1)
y, dc =pow(c,-10)

print(f"f({x0})={y}")
print(f"f'({x0})={dc*db*da*dx0}")
```
