###### 연습문제 - 3
리스트에는 집주소로 정렬된 거주인원의 이름이 저장돼 있습니다.   
`| 아름 | 종훈 | 현호 | 시후 | 아라 | 정현 |`  

----

하나의 이름을 입력받아 인접한 두 집을 출력하세요.  
`name = input("이름:")`

----

```python
# 출력 예시
>> 이름: 현호
주변에는 종훈 시후가 살고 있습니다.
```

````python
거주순서 = ["아름", "종훈", "현호", "시후", "아라", "정현"]

name = input("이름:")

if name in 거주순서:
  x=거주순서.index(name)
  print("{}에 인접한 두 집은 {},{}입니다.".format(name,거주순서[x-1],거주순서[x+1]))
````
