## 라인차트
`plot` 메서드는 라인차트를 그립니다. 첫 번째 파라미터는 X축 두 번째 파라미터는 Y축을 의미합니다.

````python
plt.plot([1, 2, 3], [100, 200, 250])
plt.show()
````

`plt.show()` 메서드가 호출되기 전에 차트에 그릴 데이터를 `plot` 메서드로 호출하면 화면에 모두 출력됩니다. 

차트제목, 축의 이름을 지정할 수 있습니다.

````python
plt.plot([1, 2, 3], [100, 200, 250])
plt.grid()
plt.xlabel("x")
plt.ylabel("y")
plt.title("title")
plt.show()
````

###### 색상 및 굵기 변경

- `linewidth` 선굵기 변경

## 막대차트
막대 차트는 연속성이 없는 여러 데이터간의 양을 비교하기에 효과적입니다. 

````python
fig, ax = plt.subplots()
ax.bar(["A", "B", "C", "D"], [100, 200, 300, 400])
 
plt.show()
````

코드를 실행하면 막대 차트가 그려집니다. bar() 메서드로 넘긴 첫 번째 파라미터(`A/B/C/D`)는 X 축 이름이 되고 두 번째 파라미터는 이에 대응되는 값(`100/200/300/400`) 입니다. 

`bar` 메서드에 `width` 옵션을 추가하면 차트의 굵기를 조절할 수 있습니다. 
- 하나의 값을 입력하면 전체 굵기가 조정됩니다. 
- 리스트로 각각의 굵기 정보를 설정할 수 있습니다.  

````python
fig, ax = plt.subplots()
ax.bar(["A", "B", "C", "D"], [100, 200, 300, 400], width=0.2)
plt.show()
````

차트의 색상은 `color` 옵션을 사용합니다. 

`bar()` 메서드는 `BarContainer` 객체를 반환하는데, 해당 객체 안에는 차트를 표현하는 `Rectangle` 객체가 들어 있습니다.  
`set_color('r')`를 사용해서 직사각형 객체의 색을 변경할 수 있습니다. 

````python
fig, ax = plt.subplots()
c = ax.bar(["A", "B", "C", "D"], [100, 200, 300, 400], width=0.2)
c[0].set_color('r')
plt.show()
````

`barh` 메서드를 사용해서 수평 방향으로 채워지는 막대 차트를 그릴 수 있습니다. 

````python
fig, ax = plt.subplots()
ax.barh(["가", "나", "다"], [10, 30, 60])
plt.show()
````

누적된 `barh` 차트를 그릴 수도 있구요.

````python
fig, ax = plt.subplots()
ax.barh(["가", "나", "다"], [10, 30, 60])
ax.barh(["가", "나", "다"], [10, 30, 60], left=[10, 30, 60])
plt.show()
````

## 서브 플롯

여러 개의 그릴 수 있는 공간을 만듭니다. 

````python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(10, 6))

x = range(0, 100)
y = [v*v for v in x]

axes[0].plot(x, y)
axes[1].bar(x, y)

plt.show()
````

## 산점도 (scatter plot)

산점도는 두 변수의 상관 관계를 직교 좌표계의 평면에 데이터를 점으로 표현하는 그래프입니다. 

````python
np.random.seed(0)    

num = 50
x = np.random.rand(num)
y = np.random.rand(num)

fig, ax = plt.subplots(figsize=(10, 6))
ax.scatter(x, y)
plt.show()
````

````python
import numpy as np
np.random.seed(0)    

# -------------------------------------------------
num = 50
x = np.random.rand(num)
y = np.random.rand(num)
c = np.random.choice([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100], num)

fig, ax = plt.subplots(figsize=(10, 6))
ax.scatter(x, y, c=c)
plt.show()
````

###### 산점도 실전 예제

````python
from pykrx import stock
````

````python
# 삼성전자 
samsung = stock.get_market_ohlcv_by_date("20200102", "20200531", "005930")
samsung.head()
````

````python
# 코스피 지수
kospi = stock.get_index_ohlcv_by_date("20200102", "20200531", "1001")
kospi.head()
````

### 축 추가하기

지수대비 상승률을 확인하고 싶어 두 개의 `라인` 차트를 한 번에 그릴 수 있습니다. 

````python
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(samsung['종가'], color='r')
ax.plot(kospi['종가'], color='g') 
plt.show()
````

두 데이터의 단위가 달라 보기 어렵게 그려졌습니다. 이러한 문제를 해결하기 위해 `ax.twinx()`는 보조 차트를 추가합니다. `twinx`가 반환하는 subplot 객체에 plot 합니다. 

````python
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(samsung['종가'], color='r')

# --------------------- NEW ----------------------
sub = ax.twinx()
# ------------------------------------------------
sub.plot(kospi['종가'], color='g') 
plt.grid()
plt.show()
````
