# 반복문
반복문은 자료구조의 데이터 수만큼 코드를 반복실행 합니다.  
자료구조에 있는 "데이터를 하나씩 바인딩하고 들여쓰기된 코드를 실행" 과정을 반복하는 겁니다. 
  
## 반복문 - 기본형
`for 변수 in 자료구조 : ` 형태로 코드를 작성합니다. 코드 끝에 콜론 `:` 있음에 주의하세요. 분기문과 같이 콜론 다음에는 들여쓰기와 함께 영향을 받는 코드를 작성합니다. 

<img src="https://i.ibb.co/gR952tp/for-0.png" width="300" style="float:left" />

###### 반복문을 사용하지 않고 한 줄에 하나의 값 출력하기

````python
리스트 = ['가','나','다','라']
print(리스트[0])
print(리스트[1])
print(리스트[2])
print(리스트[3])
````

###### 반복문으로 코드를 정리하기

````python
for a in 리스트:
  print(a)
````

````python
리스트 = ['철수','영희','희철']

for name in 리스트:
  print(name)
````

````python
list_name=['동희','호로록']

for name in list_name:
  print(f"{name} 바보")
````

````python
n =10
list1=[1,2,3,4,5]
for i in list1:
    print(n)
````

###### 연습문제 - 2
리스트에서 앞의 두 개의 값을 다음과 같이 출력하라.  
```
가
나
```

````python
# 슬라이싱 이용해서 문제를 해결하라.
리스트 = ["가", "나", "다"]
````

````python
for i in 리스트[:2]: #리스트를 슬라이싱으로 재구성
    print(i)
````

````python
# 분기문을 이용해서 문제를 해결하라.
리스트 = ["가", "나", "다"]
````

````python
for i in 리스트:
    if i =='다':
        break
    print(i)
    
````

###### 연습문제 - 3
20 이상인 값을 출력하라

````python
리스트 = [20, 10, 30]
for 변수 in 리스트:
    if 변수 >= 20:
        print(변수)
````

20 이상인 값을 저장하라

````python
리스트2 =[]
for 변수 in 리스트:
    if 변수>=20:
        리스트2.append(변수)

리스트2
````
